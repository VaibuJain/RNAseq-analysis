## **Normalized tpm expression**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the normalized data
# Update the file path as necessary
df = pd.read_csv('/content/drive/normalized_abundance.csv')

# Drop rows with NaN values in relevant columns to avoid issues in plotting
df_cleaned = df.dropna(subset=['Coverage', 'FPKM', 'TPM'])

# Set gene_id as index for heatmap
df_heatmap = df_cleaned.set_index('Gene ID')[['Coverage', 'FPKM', 'TPM']]

# Step 2: Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_heatmap, cmap="RdYlBu_r", cbar_kws={'label': 'Normalized Value'})
plt.title("Heatmap of Normalized lncRNA Metrics", fontsize=25)  # Increase title font size
plt.xlabel("Metrics", fontsize=25)  # Corrected x-label to "Metrics"
plt.ylabel("Gene ID", fontsize=25)  # Corrected y-label to "Gene ID"
plt.xticks(fontsize=20)  # Increase font size for x-ticks
plt.yticks(fontsize=20)  # Increase font size for y-ticks
plt.show()

# Step 3: Boxplot for each metric
plt.figure(figsize=(10, 6))
# Define custom color palette for boxplots
boxplot_palette = {
    'Coverage': '#EE4E4E',  # Custom color for Coverage
    'FPKM': '#FFF455',      # Custom color for FPKM
    'TPM': '#219C90'        # Custom color for TPM
}
sns.boxplot(data=df_cleaned[['Coverage', 'FPKM', 'TPM']], palette=boxplot_palette)
plt.title("Boxplot of Normalized lncRNA Metrics", fontsize=25)  # Increase title font size
plt.ylabel("Metrics", fontsize=25)  # Increase y-label font size
plt.xlabel("Metrics", fontsize=25)  # Increase x-label font size
plt.xticks(fontsize=20)  # Increase font size for x-ticks
plt.yticks(fontsize=20)  # Increase font size for y-ticks
plt.show()
```
